```python
import pandas as pd
import sklearn
from sklearn.model_selection  import train_test_split 
```

```python
df=pd.read_csv("NY_house.csv")
df.head()
```

```python
df.isna().sum()
```

```python
df.nunique()
```

```python
df.drop(columns=['ADDRESS','MAIN_ADDRESS','FORMATTED_ADDRESS','LATITUDE','LONGITUDE','LONG_NAME','BROKERTITLE'],inplace=True)
```

```python
df.head()
df.nunique()
```

```python
df.info()
```

```python
df.head()
```

```python
df=pd.get_dummies(df)
```

```python
df.corr()
```

```python
x_df=df.drop(columns=['PRICE'])
y_df=df['PRICE']
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_df,y_df,test_size=0.25,random_state=2)
```

```python
model.fit(x_train,y_train)
```

```python
prediction=model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.linear_model import Lasso
model2=Lasso(alpha=50,max_iter=100,tol=0.1)
```

```python
model2.fit(x_train,y_train)
```

```python
model2.predict(x_test)
```

```python
model2.score(x_test,y_test)
```
